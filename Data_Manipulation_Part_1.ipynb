# **Data Manipulation**

##Import Library Pandas

```python
import pandas as pd
```

##Membaca Dataset

### Membaca dataset CSV dan TSV

```python
# File CSV
df_csv = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/sample_csv.csv")
df_csv
```

```python
# File TSV
df_tsv = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/sample_tsv.tsv",sep='\t')
df_tsv
```

###Membaca dataset file Excel

```python
#File Excel
df_excel = pd.read_excel("https://storage.googleapis.com/dqlab-dataset/sample_excel.xlsx")
df_excel
```

### Membaca File Dataset JSON

```python
#File JSON
url = "https://storage.googleapis.com/dqlab-dataset/covid2019-api-herokuapp-v2.json"
df_json = pd.read_json(url)
df_json
```

##Menyimpan Dataset

```python
# Menyimpan dataset ke bentuk CSV
df_csv.to_csv("csv1.csv", index=False)
```

```python
# Menyimpan dataset ke bentuk TSV
df_tsv.to_csv("tsv1.tsv", index=False, sep='\t')
```

```python
# Menyimpan dataset ke bentuk Excel
df_excel.to_excel("xlsx1.xlsx", index=False)
```
